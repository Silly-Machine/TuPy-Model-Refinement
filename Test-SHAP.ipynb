```python
import shap
import torch
from transformers import  pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
#Sets
model_path='Silly-Machine/TuPy-Bert-Large-Binary-Classifier'

# Enable GPU
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

# Define the tokenizer (replace 'bert-base-multilingual-cased' with your desired tokenizer)
tokenizer = AutoTokenizer.from_pretrained(model_path)

# Load the pre-trained model and move it to the device
loaded_model = AutoModelForSequenceClassification.from_pretrained(model_path).to(device)
loaded_model.eval()
```

```python
#examples

examples = ['olá,bom dia',
            "bom dia,vadia gorda🤣" ,
            'Pra ser minha inimiga no mínimo tem que ter um rostinho bonito e delicado, não se considere minha rival com essa sua cara de cavalo não, feia, cara de traveco, cabeçuda, queixo quadrado 🤣🤣'
            ]

# build a pipeline object to do predictions
classifier = pipeline(
    "text-classification",
    model=loaded_model,
    tokenizer=tokenizer,
    device=device,
    return_all_scores=True,
)
#
explainer = shap.Explainer(classifier,tokenizer)
shap_values = explainer(examples)
#
try:
  shap_values_hate = shap_values[:, :, "hate"]
  shap_values_not_hate = shap_values[:, :, "not hate"]
  shap.plots.text(shap_values_hate)
  shap.plots.text(shap_values_not_hate)
except:
   shap.plots.text(shap_values)
```
